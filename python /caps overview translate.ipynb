```python
def translate_text(target, text):
    import six
    from google.cloud import translate_v2 as translate

    translate_client = translate.Client()

    if isinstance(text, six.binary_type):
        text = text.decode("utf-8")

    result = translate_client.translate(text, target_language=target)

    #print(u"Text: {}".format(result["input"]))
    #print(u"Translation: {}".format(result["translatedText"]))
    #print(u"Detected source language: {}".format(result["detectedSourceLanguage"]))

    return result["translatedText"]
```

```python
for result in results:
    contentid = result['contentid']
    print(contentid)
    #overview_jap = translate_text("ja", result['overview_eng'])
    overview_kor = translate_text("ko", result['overview_eng'])
    print(overview_kor)
    insert_query = "INSERT IGNORE INTO overview_tb(contentid, overview_kor) VALUES(%s, %s)"
    values = (str(contentid), overview_kor)
    cursor.execute(insert_query, values)
    db.commit()
    print("insert ok")
    
```
